```python
import requests
import json
import datetime

# --- 設定 ---
BASE_URL = "https://protrainer-dev.github.io/protrainer-contents"
START_YEAR = 2026
START_MONTH = 3
FILE_NAME = "sentences_l18n.json"
OUTPUT_FILE = "merged_sentences_master.json"

def fetch_and_merge():
    all_sentences = []
    
    # 現在の年月を取得
    now = datetime.datetime.now()
    current_year = now.year
    current_month = now.month

    print(f"🚀 {START_YEAR}{START_MONTH:02d} からのデータ統合を開始します...")

    # ループ処理
    year = START_YEAR
    month = START_MONTH
    
    while (year < current_year) or (year == current_year and month <= current_month):
        # YYYYMM 形式の文字列を作成
        date_str = f"{year}{month:02d}"
        url = f"{BASE_URL}/{date_str}/{FILE_NAME}"
        
        print(f"🔍 チェック中: {url}")
        
        try:
            response = requests.get(url)
            if response.status_code == 200:
                data = response.json()
                # データがリスト形式であることを想定
                if isinstance(data, list):
                    all_sentences.extend(data)
                    print(f"   ✅ {len(data)} 件のフレーズを追加しました。")
                else:
                    print(f"   ⚠️ 警告: {date_str} の形式がリストではありません。")
            elif response.status_code == 404:
                print(f"   💨 スキップ: ファイルが存在しません。")
            else:
                print(f"   ❌ エラー: ステータスコード {response.status_code}")
        except Exception as e:
            print(f"   ❌ 通信エラー: {e}")

        # 月を進める
        month += 1
        if month > 12:
            month = 1
            year += 1

    # 重複IDの除去（念のため）
    unique_sentences = list({s['id']: s for s in all_sentences}.values())
    
    # ファイル保存
    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(unique_sentences, f, ensure_ascii=False, indent=2)

    print("\n" + "="*40)
    print(f"✅ 完了！")
    print(f"📊 統合後の総フレーズ数: {len(unique_sentences)} 件")
    print(f"💾 保存先: {OUTPUT_FILE}")
    print("="*40)

if __name__ == "__main__":
    fetch_and_merge()
```
